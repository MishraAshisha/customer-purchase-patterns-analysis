```python
#optimising column names for easier uses
df.columns = df.columns.str.lower()
df.columns =  df.columns.str.replace(" ","_")
df.columns
```

```python
#renaming a column to make it easier to read
df = df.rename( columns = {"purchase_amount_(usd)" : "purchase_amount"})
df.columns
```

```python
#checking for null values
df.isnull().sum()
```

```python
#using median() to fill up null values 
df["review_rating"] = df.groupby("category")["review_rating"].transform(lambda x : x.fillna( x.median()))
df["review_rating"] 
```

```python
#checking if any null value is left after using the median()
df.isnull().sum()
```

```python
#creating labels for the new column - "age_groups" that we are creating by grouping the age numbers in the column - "ages".
labels = ["young_adult" , "adult" , "middle_age" , "senior"]
#creating the new column - "age_groups" which separates the already existing column - "ages"
df["age_groups"] = pd.qcut(df["age"],q=4, labels=labels)
#checking the separation 
df[["age_groups", "age"]]. head(10)
```

```python
#checking the values in the column 
df["frequency_of_purchases"].unique()
```

```python
#after checking we use map() to link a numerical value to the text values
df["days_between_puchases"] = df["frequency_of_purchases"]. map({
    "Fortnightly": 14, 
    "Weekly": 7, 
    'Annually': 365, 
    'Quarterly':90, 
    'Bi-Weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
})

#checking if the numerical values were correctly linked to the text value
df[["days_between_puchases" , "frequency_of_purchases"]].head(10)
```
